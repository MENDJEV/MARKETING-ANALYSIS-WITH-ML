```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler

```

```python
df = pd.read_csv('../data/raw/marketing_campaign.csv', sep='\t')
df
```

```python
inf_cols = df.isin([np.inf, -np.inf]).sum()
print(inf_cols[inf_cols > 0])
```

```python
df['Education'].unique()
df['Marital_Status'].unique()
```

```python
#Suppression des colonnes inutiles
df.drop(['ID','Z_CostContact','Z_Revenue'],axis = 1, inplace = True)
df
```

```python
#Remplaçons year_birth par age
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')
last_enrollment = df['Dt_Customer'].max()
df['Age'] = last_enrollment.year - df['Year_Birth']
df.drop('Year_Birth', axis=1, inplace=True)
df['Nb_Jours_Client'] = (last_enrollment - df['Dt_Customer']).dt.days
df.drop('Dt_Customer', axis=1, inplace=True)
df
```

```python
#création de la colonne Total_depenses
df['Total_depenses'] = (df['MntWines'] + df['MntFruits'] +
                        df['MntMeatProducts'] + df['MntFishProducts'] +
                        df['MntSweetProducts'] + df['MntGoldProds'])
df
```

```python
#Encodage de education
df['Education'] = df['Education'].str.strip()
education_mapping = {
    'Basic'      : 0,
    '2n Cycle'   : 1,
    'Graduation' : 2,
    'Master'     : 3,
    'PhD'        : 4
}
df['Education'] = df['Education'].map(education_mapping)
```

```python
#Encodage de Marital_Status
df['Marital_Status'] = df['Marital_Status'].str.strip()
marital_mapping = {
    'Single'   : 0,
    'Divorced' : 0,
    'Widow'    : 0,
    'Alone'    : 0,
    'Absurd'   : 0,
    'YOLO'     : 0,
    'Together' : 1,
    'Married'  : 1
}
df['Marital_Status'] = df['Marital_Status'].map(marital_mapping)
```

```python
#Suppression des valeurs manquantes
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
#création de la feature Total_kids
df['Total_kids'] = df['Kidhome'] + df['Teenhome']
```

```python
#création de la feature Total_purchases
df['Total_purchases'] = df['NumWebPurchases'] +df['NumStorePurchases']  + df['NumCatalogPurchases']
+ df['NumDealsPurchases']
```

```python
#création des features depenses_moyenne_par_achat et tauc_de_reponse_aux_campagnes
df['Avg_spend_per_purchase'] = df.apply(
    lambda row: row['Total_depenses'] / row['Total_purchases'] 
    if row['Total_purchases'] > 0 else 0, axis=1
)
campagnes = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 
             'AcceptedCmp4', 'AcceptedCmp5', 'Response']

df['Campaign_Response_Rate'] = df[campagnes].sum(axis=1) / len(campagnes)
```

```python
#Suppression des colonnes Campagnes ainsi que les features Kidhome et Teenhome
campagnes = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 
             'AcceptedCmp4', 'AcceptedCmp5', 'Response']
df.drop(campagnes,axis = 1, inplace = True)
df.drop(['Kidhome','Teenhome'],axis = 1, inplace = True)
```

```python
df.to_csv('../data/processed/clean_marketing_campaign.csv', index=False)
```

```python
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
df_scaled.to_csv('../data/processed/scaled_marketing_campaign.csv', index=False)
```
