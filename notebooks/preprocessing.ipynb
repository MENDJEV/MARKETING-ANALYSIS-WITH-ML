```python
#1.iMPORTATION DES BIBLIOTHÈQUES

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#2. CHARGEMENT DU JEU DE DONNÉES

df = pd.read_csv("../data/raw/marketing_campaign.csv", sep="\t")
df
```

```python
#3.ENCODAGE EDUCATION ET SITUATION MARITALE

#encodage education

education_map = {
    'Basic'      : 0,
    '2n Cycle'   : 1,
    'Graduation' : 2,
    'Master'     : 3,
    'PhD'        : 4
}

df['Education'] =df['Education'].str.strip().map(education_map)

#encodage marital_status
marital_map = {
    'Single'   : 0,
    'Together' : 1,
    'Married'  : 1,
    'Divorced' : 0,
    'Widow'    : 0,
    'Alone'    : 0,
    'Absurd'   : 0,
    'YOLO'     : 0
}
df['Marital_Status']=df['Marital_Status'].map(marital_map)
df.shape
```

```python
#4. CREATION DE LA COLONNE SOMME DEPENSES

df['Somme_depenses'] =df[['MntWines','MntFruits','MntMeatProducts','MntSweetProducts','MntGoldProds']].sum(axis=1)
print(df.shape)
df
```

```python
#5. CREATION DE LA COLONNE AGE

# Format jour-mois-année
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')

# Extraire la date max
date_max = df['Dt_Customer'].max()

# Calculer l'âge
df['Year_Birth'] = date_max.year - df['Year_Birth']

print(df.shape)
df.head()
```

```python
#6. SUPPRESSION DES COLONNES INUTILES POUR NOTRE ANALYSE

#a.collection des différentes colonnes à supprimer
cols_a_supprimer = [

    'ID',            # identifiant non informatif
    'Z_Revenue',     # colonne constante inutile
    'Z_CostContact', # colonne constante inutile
]

#b.suppression effective de ces colonnes
df.drop(columns=cols_a_supprimer, inplace=True,errors='ignore')

#c.affichage et vérification
print(df.shape)
```

```python
#7. CONVERSION DE DT_CUSTOMER EN NOMBRE DE JOURS DEPUIS L'INSCRIPTION

df['Dt_Customer'] =(date_max - df['Dt_Customer']).dt.days
print(df.shape)
df
```
