```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#valeurs manquantes
missing_values = df.isna().sum()
missing_values
```

```python
# Visualisation de la distribution des variables
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')
last_enrollment = df['Dt_Customer'].max()
df['Age'] = last_enrollment.year - df['Year_Birth']
df['Total_depenses'] = (df['MntWines'] + df['MntFruits'] + 
                        df['MntMeatProducts'] + df['MntFishProducts'] + 
                        df['MntSweetProducts'] + df['MntGoldProds'])

features = ['Age', 'Income', 'Total_depenses', 
            'MntWines', 'MntMeatProducts', 
            'NumWebPurchases', 'NumStorePurchases', 
            'NumDealsPurchases']

fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()

colors = ['steelblue', 'coral', 'mediumseagreen', 
          'mediumpurple', 'orange', 'crimson', 
          'teal', 'goldenrod']

for i, col in enumerate(features):
    n_missing = df[col].isnull().sum()
    pct_missing = (n_missing / len(df)) * 100
    
    # Histogramme SANS dropna()
    axes[i].hist(df[col], bins=30, 
                 color=colors[i], edgecolor='white', alpha=0.8)
    
    # Titre avec info sur les manquants
    if n_missing > 0:
        axes[i].set_title(f'Distribution : {col}\n {n_missing} manquants ({pct_missing:.1f}%)', 
                          fontsize=11, fontweight='bold', color='crimson')
    else:
        axes[i].set_title(f'Distribution : {col}\n 0 manquant', 
                          fontsize=11, fontweight='bold')
    
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Fréquence')
    
    # Ligne de moyenne
    mean_val = df[col].mean()
    axes[i].axvline(mean_val, color='red', linestyle='--', linewidth=1.5)
    axes[i].text(mean_val, axes[i].get_ylim()[1]*0.9, 
                 f'Moy: {mean_val:.0f}', color='red', fontsize=9)

axes[-1].set_visible(False)
plt.suptitle('Distribution des variables clés', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Analyse de la corrélation - features clés uniquement
marital_mapping = {
    'Single'   : 1,
    'Together' : 2,
    'Married'  : 3,
    'Divorced' : 4,
    'Widow'    : 5,
    'Alone'    : 1,
    'Absurd'   : 0,
    'YOLO'     : 0
}
df['Marital_Encoded'] = df['Marital_Status'].map(marital_mapping)

features_corr = ['Age', 'Income', 'Total_depenses',
                 'MntWines', 'MntMeatProducts', 'MntFruits',
                 'MntFishProducts', 'MntSweetProducts',
                 'NumWebPurchases', 'NumStorePurchases', 'NumCatalogPurchases',
                 'NumDealsPurchases', 'NumWebVisitsMonth','Marital_Encoded']


plt.figure(figsize=(12, 8))
sns.heatmap(df[features_corr].corr(), 
            annot=True, 
            fmt='.2f', 
            cmap='coolwarm', 
            center=0,
            square=True,
            linewidths=0.5)

plt.title('Heatmap des corrélations - Variables clés', fontsize=16, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Définition des paires
pairs = [
    ('Total_depenses', 'MntWines',           0.89),
    ('Total_depenses', 'MntMeatProducts',    0.84),
    ('Income',         'Total_depenses',     0.67),
    ('MntFruits',      'MntFishProducts',    0.59),
    ('Income',         'NumStorePurchases',  0.53),
    ('Income',         'NumWebVisitsMonth', -0.55),
    ('MntMeatProducts','NumWebVisitsMonth', -0.54),
]

fig, axes = plt.subplots(3, 3, figsize=(18, 15))
axes = axes.flatten()

colors = ['steelblue', 'coral', 'mediumseagreen', 
          'mediumpurple', 'orange', 'crimson', 'teal']

for i, (x, y, corr) in enumerate(pairs):
    sns.scatterplot(data=df, x=x, y=y, 
                    alpha=0.4, color=colors[i], ax=axes[i])
  
    
    # Couleur du titre selon corrélation
    title_color = 'crimson' if corr > 0 else 'steelblue'
    axes[i].set_title(f'{x} vs {y}\ncorrélation = {corr}', 
                      fontsize=11, fontweight='bold', color=title_color)
    axes[i].set_xlabel(x)
    axes[i].set_ylabel(y)

# Cacher les subplots vides
for j in range(len(pairs), len(axes)):
    axes[j].set_visible(False)

plt.suptitle('Relations entre variables clés', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
