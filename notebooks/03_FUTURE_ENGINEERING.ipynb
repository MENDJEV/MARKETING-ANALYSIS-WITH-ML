```python
"""
Future Engineering - Customer Personality Analysis
Input : data/processed/df_clean.csv
Output: data/processed/df_features.csv
"""

import pandas as pd
import os
```

```python
# Chargement des données nettoyées
DATA_PATH = "../data/processed/df_clean.csv"
df_clean = pd.read_csv(DATA_PATH)

# Copie de travail
df_feat = df_clean.copy()

print(f"Jeu de données initial: {df_feat.shape}")
```

```python
# --- Choix de la date de référence ---

df_feat["Dt_Customer"] = pd.to_datetime(df_feat["Dt_Customer"], dayfirst=True)

ref_date = df_feat["Dt_Customer"].max()
ref_year = ref_date.year


print(f"Date de référence: {ref_date}")
print(f"Année de référence: {ref_year}")
print(df_feat["Dt_Customer"].dtype)
```

```python
# --- 1. Variable Age ---
# Transformation de la variable Year_Birth en variable Age

df_feat["Age"] = ref_year - df_feat["Year_Birth"]

print(f"Age min : {df_feat["Age"].min()}")
print(f"Age max : {df_feat["Age"].max()}")
print(f"Age moyen : {df_feat["Age"].mean().round(1)}")
```

```python
# --- 2. Variable Ancienneté ---
# Nombre de jours entre l'inscription d'un client et la date de référence


df_feat["Seniority"] = (ref_date - df_feat["Dt_Customer"]).dt.days

print(f"Ancienneté min : {df_feat["Seniority"].min()} jours ")
print(f"Ancienneté max : {df_feat["Seniority"].max()} jours ")
print(f"Ancienneté moyenne : {df_feat["Seniority"].mean().round(1)} jours ")
```

```python
# --- 3. Variable Total dépenses ---
# Somme de toutes les catégories de produits sur 2 ans

spend_cols = [
    "MntWines", "MntFruits", "MntMeatProducts",
    "MntFishProducts", "MntSweetProducts", "MntGoldProds"
]

df_feat["Total_spend"] = df_feat[spend_cols].sum(axis=1)

print(f"Total dépenses min: {df_feat["Total_spend"].min()}")
print(f"Total dépenses max: {df_feat["Total_spend"].max()}")
print(f"Total dépenses min: {df_feat["Total_spend"].mean().round(2)}")
```

```python
# --- 4. Variable Total Campagnes
# Nombre de campagnes marketing acceptées sur les 6 disponibles

campaign_cols = [
    "AcceptedCmp1", "AcceptedCmp2", "AcceptedCmp3",
    "AcceptedCmp4", "AcceptedCmp5", "Response"
]

df_feat["Total_campaign"] = df_feat[campaign_cols].sum(axis=1)

print("Distribution totale des campagnes")
print(df_feat["Total_campaign"].value_counts().sort_index())
```

```python
# --- 5. Variable nombre d'enfant ---
# Nombre total d'enfants dans le foyer (enfants + adolescents)

df_feat["Nb_kids"] = df_feat["Kidhome"] + df_feat["Teenhome"]

print("Distribution Nb_Enfants:")
print(df_feat["Nb_kids"].value_counts().sort_index())
```

```python
# --- 6. Nettoyage final ---
# Suppression des colonnes source remplacées par les features créées

cols_to_drop = [
    "Year_Birth",
    "Dt_Customer",
    "Kidhome",
    "Teenhome",
]

df_feat = df_feat.drop(columns=cols_to_drop)


print("=" * 50)
print("RAPPORT FEATURE ENGINEERING")
print("=" * 50)
print(f"Jeu de données initial initial   : {df_clean.shape}")
print(f"Jeu de données final     : {df_feat.shape}")
print(f"Features créées : Age, Seniority, Total_spend,")
print(f"                  Total_campaign, Nb_kids")
print(f"Colonnes restantes :")
print(df_feat.columns.tolist())
print("=" * 50)


# Sauvegarde
output_path = "../data/processed"
os.makedirs(output_path, exist_ok=True)
df_feat.to_csv(f"{output_path}/df_features.csv", index=False)

print(f"Sauvegarde réussi : {output_path}/df_features.csv")
```
