```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
df = pd.read_csv("../data/processed/scaled_marketing_campaign.csv")
```

```python
df.head()
```

```python
Cah = linkage(df,method='ward',metric='euclidean')
```

```python
plt.figure(figsize=(14, 7))
dendrogram(
    Cah,
    truncate_mode='lastp',
    p=30,
    color_threshold=0.7 * max(Cah[:, 2]),
    leaf_rotation=45,
    leaf_font_size=10,
    show_contracted=True
)

# Ligne de coupe
seuil = 0.7 * max(Cah[:, 2])
plt.axhline(y=seuil, color='red', linestyle='--', 
            linewidth=2, label=f'Seuil de coupe = {seuil:.1f}')

plt.title('Dendrogramme — CAH (méthode Ward)', 
          fontsize=14, fontweight='bold')
plt.xlabel('Clients')
plt.ylabel('Distance euclidienne')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
