```python
"""
Preprocessing - Customer Personality Analysis
Input : data/raw/marketing_campaign.csv
Output: data/processed/df_clean.csv
"""

import pandas as pd
import os

```

```python
# Chargement des données brutes

DATA_PATH = "../data/raw/marketing_campaign.csv"


df = pd.read_csv(DATA_PATH, sep="\t")

# Copie du travail
df_clean = df.copy()

print(f"Copie du jeu de données initial {df_clean.shape}")
```

```python
# --- 1. Suppression des colonnes inutiles ---

# On recherche calcule la variance de chaque variable
var_numeriques = df_clean.var(numeric_only=True)

# On récupère les colonnes dont la variance est nulle
cols_variance_nulle = var_numeriques[var_numeriques == 0].index.tolist()

print(f"Colonnes à variance nulle : {cols_variance_nulle}")

cols_a_supprimer = ["ID"] + cols_variance_nulle
df_clean = df_clean.drop(columns=cols_a_supprimer)

print(f"Colonnes supprimées : {cols_a_supprimer}")
print(f"Shape après suppression : {df_clean.shape}")
```

```python
# --- 2. Gestion des valeurs manquantes ---

# Vérifications
colonnes_val_nulles = [x for x in df_clean.isnull().sum()[df_clean.isnull().sum() > 0].index]


print(f"colonnes à valeurs manquantes avant :{colonnes_val_nulles}")

# Suppression des lignes avec NaN
df_clean = df_clean.dropna(subset=colonnes_val_nulles)

print(f"Colonnes à valeurs nulles après suppression:  {df_clean.isnull().sum()[df_clean.isnull().sum() > 0]}")
print(f"Jeu de données apres suppression des valeurs nulles {df_clean.shape}")
```

```python
# --- 3. Suppression des valeurs abérrantes

print(f"Shape avant nettoyage des valeurs abérrantes: {df_clean.shape}")

# Valeurs abérrantes de la variable Birth_Year
min_year = 1930
len_before = len(df_clean)
df_clean = df_clean[df_clean["Year_Birth"] >= min_year]
len_after = len(df_clean)

print(f"Year_Birth < {min_year}: {len_before - len_after} ligne(s) supprimée(s)")
print(f"Valeurs supprimées: {df['Year_Birth'][df['Year_Birth'] < min_year].values}")


# Valeurs abérrantes de la variable Income
max_income = 200000
len_before = len(df_clean)
df_clean = df_clean[df_clean["Income"] <= max_income]
len_after = len(df_clean)

print(f"\n Income > {max_income}: {len_before - len_after} ligne(s) supprimée(s)")
print(f"Valeurs supprimées: {df['Income'][df['Income'] > max_income].values}")
print(f"\n Jeu de données après nettoyage des valeurs abérrantes: {df_clean.shape}")
```

```python
# --- 4. Nettoyage des variables qualitatives ---

print(df_clean["Marital_Status"].value_counts())

marital_mapping = {
    "Alone": "Single",
    "YOLO": "Single",
    "Absurd": "Single",
}

df_clean["Marital_Status"] = df_clean["Marital_Status"].replace(marital_mapping)

print(df_clean["Marital_Status"].value_counts())
print(df_clean.shape)
```

```python
# --- 5. Vérification finale et sauvegarde

print("=" * 50)
print("Rapport de nettoyage")
print("=" * 50)
print(f"Lignes initiales: {len(df)}")
print(f"Lignes finales: {len(df_clean)}")
print(f"Lignes supprimées: {len(df) - len(df_clean)}")
print(f"Colonnes initiales: {df.shape[1]}")
print(f"Colonnes finales: {df_clean.shape[1]}")
print(f"Valeurs manquantes: {df_clean.isnull().sum().sum()}")
print("="*50)


# Vérification des variables qualitatives nettoyées
print("\n Marital_Status - valeurs uniques restantes:")
print(df_clean["Marital_Status"].unique())


# Sauvegarde
output_path = "../data/processed"
os.makedirs(output_path, exist_ok=True)
df_clean.to_csv(f"{output_path}/df_clean.csv", index=False)

print(f"\n Jeu de données propre sauvegardé {output_path}/df_clean.csv")
```
